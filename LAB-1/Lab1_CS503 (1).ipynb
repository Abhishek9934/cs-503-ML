```
# Normalizing the dataset using MinMaxScaler from sklearn library

min_max_scaler = preprocessing.MinMaxScaler()
normalised_data = min_max_scaler.fit_transform(iris.data)


# Print dataset before normalization          
print(iris.data)

# Print dataset after normalization  
print(normalised_data)


normalised_df = pd.DataFrame(normalised_data)
normalised_df.columns = iris.feature_names
normalised_df.head()
```

##Inferences
The Normalisation had changed the values of the attributes to fit in the range of [0 , 1]. In this approach, for each featurem the minimum data value is changed to 0 and the maximum value is changed to 1 , the rest of the values gets changed between 0 and 1. In the histogram it can be seen all the values are between 0 and 1. This helps in carrying out operations on the dataset with more ease.

```
# For entire normalized dataset, plot 4 histograms for sepal length (cm), sepal width (cm), petal length (cm) and petal width (cm)


plotHistogram(normalised_df['sepal length (cm)'] ,'sepal length (cm)' ,'green' )
plotHistogram(normalised_df['sepal width (cm)'] ,'sepal width (cm)' ,'green')
plotHistogram(normalised_df['petal length (cm)'] ,'petal length (cm)' ,'green')
plotHistogram(normalised_df['petal width (cm)'] ,'petal width (cm)' ,'green')
```

```
# Data set contains samples 0:50 of class 0, 50:100 of class 1, 100:150 of class 2

#creating separate dataframe for each class of iris dataset

setosa = pd.DataFrame(iris.data[0:50]) 
versicolor = pd.DataFrame(iris.data[50:100])
virginica = pd.DataFrame(iris.data[100:])


# For entire dataset, plot 4 graphs for sepal length (cm), sepal width (cm), petal length (cm) and petal width (cm).

# Function to generate the graph 
def plotGraph( x, y , z , xlabel):
  plt.figure(figsize=(20,7))
  plt.subplot(1,3 ,1)
  plt.hist(x  , color = 'blue' , label = 'setosa')
  plt.xlabel(xlabel)
  plt.ylabel('frequency')
  plt.legend()
  plt.subplot(1,3,2)
  plt.hist(y  , color = 'orange' , label = 'versicolor')
  plt.xlabel(xlabel)
  plt.ylabel('frequency')
  plt.legend()
  plt.subplot(1,3,3)
  plt.hist(z , color = 'green' , label = 'virginica' )
  plt.xlabel(xlabel)
  plt.ylabel('frequency')
  plt.legend()
  plt.show()


# Each graph will have three differently colored (3 colors for 3 classes) histograms.

plotGraph(setosa[0] ,versicolor[0],virginica[0],'sepal length (cm)' )
plotGraph(setosa[1] ,versicolor[1],virginica[1],'sepal width (cm)' )
plotGraph(setosa[2] ,versicolor[2],virginica[2],'petal length (cm)' )
plotGraph(setosa[3] ,versicolor[3],virginica[3],'petal width (cm)' )


```

Let's study the relationship between attributes.
1. Find Pearson correlation for each pair of attributes. {Expected output is a table (triangular matrix)} **(2 points)** 
2. Also, draw a heatmap representing this correlation. **(2 points)** 

```
# Compute Pearson correlation between each pair of attributes
iris_data = pd.DataFrame(iris.data)
iris_data.columns = iris.feature_names 

# calculating the Pearson correlation
corr = iris_data.corr(method= 'pearson') 

corr = pd.DataFrame(corr).to_numpy()  #converting the corr dataframe to numpy


# The table is going to be symmetric, so you can make a triangular matrix/table 
print("\nThe correlation matrix which gets formed is :\n")

print(corr)


#np.tril - returns the lower part of a matrix

triangular_matrix = np.tril(corr)


print("\nThe triangular matrix formed is :\n")

print(triangular_matrix)

display(pd.DataFrame(triangular_matrix))


# Generate a heatmap according to the person coefficient values 
# using seaborn library to generate the heat map

print("\n****************Heat Map*****************\n")
sns.heatmap(triangular_matrix, cmap=sns.diverging_palette(400,12 , as_cmap=True), xticklabels= iris.feature_names, yticklabels= iris.feature_names)
```
